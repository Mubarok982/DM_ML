```python
!pip install apyori
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
data = pd.read_excel("dataset.xlsx")
data.info()
```

```python
data.head()
```

```python
data = data.drop(['Tanggal', 'ID Transaksi'],axis=1)
data.head()
```

```python
# Membuat list dalam list dari transaksi pembelian barang
records = []
for i in range(data.shape[0]):
    row_items = []
    for j in range(data.shape[1]):
        cell = str(data.values[i, j])
        if cell != 'nan':
            items = cell.split(',')
            row_items.extend([item.strip() for item in items])
    records.append(row_items)
```

```python
#Menggunakan fungsi apriori untuk membuat asosiasi
association_rules = apriori(records, min_support=0.14, min_confidence=0.80, min_lift=1)
association_results = list(association_rules)
```

```python
# Menampilkan hasil perhitungan asosiasi dari item
pd.set_option('max_colwidth', 1000)
results_list = []
for item in association_results:
    for i in item.ordered_statistics:
        if i.lift != 1:
            rule_str = ', '.join(i.items_base) + ' -> ' + ', '.join(i.items_add)
            results_list.append({
                'Rule': rule_str,
                'Support': f"{round(item.support * 100, 2)}%",
                'Confidence': f"{round(i.confidence * 100, 2)}%"
            })

# Ubah ke DataFrame sekaligus
Result = pd.DataFrame(results_list)
print(Result)
```
